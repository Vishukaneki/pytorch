``` RESHAPING , STACKING , SQUEEZING AND UNSQEEZING```

````python
# Stacking torch.vstack,torch.hstack for merging stacking tensors
# view return a view of an input tensor of certain shape but keep the same
#memory as the original tensor
````

```squeeze removes all `1` Dimsension from tensor
torch.unsqueeze add 1 d to target tensor
permute return a view of the input with dimensions permuted (swapped) in a certain way```

````python
import torch
x= torch.arange(1.,10.)
x, x.shape
````

````python
x_reshaped=x.reshape(9,1)
x_reshaped,x_reshaped.shape
````

````python
z=x.view(1,9)
z,z.shape
# x and z share same memeory so change in z wll reflect change in x
````

````python
# stackng time
x_stacked=torch.stack([x,x,x,x,x,x,x])
x_hstacked=torch.hstack([x,x,x,x])
x_stacked,x_hstacked
````

````python
# time to squeeze
x= torch.zeros(2,1,2,1)
x_squeezed=x.squeeze()
x_squeezed #only 2,2 shape left
# it removes all 1 d's
````

````python
# time to unsqueeze adds a single dim to target tensor at required place well
x_unsqueezed=x_squeezed.unsqueeze(dim=2)
x_unsqueezed.shape
````

````python
torch.permute(x_unsqueezed,(1,0,2))
# used with images well
x_original=torch.rand(size=(224,224,3)) #height width color_channels
x_permuted=x_original.permute(2,0,1) # shifts axis
print(x_original.shape)
print(x_permuted.shape)
# a view shares same memory with the original so changes will be visible both sides
#return a view of the oriinal input with its dimensions permuted
````
